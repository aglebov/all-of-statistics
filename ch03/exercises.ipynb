#### Exercise 1

Denote by $i$ the number of trials in which the amount doubled. The final amount after $n$ trials is then $2^{2i - n}c$ and the probability of such an outcome is ${n \choose i}(1/2)^n$.

The expectation therefore is
$$\sum_{i=0}^n 2^{2i - n}c {n \choose i}\left(\frac12\right)^n 
= \frac{c}{2^{2n}} \sum_{i=0}^n {n \choose i} 2^{2i}
= \frac{c}{4^n} \sum_{i=0}^n {n \choose i} 4^{i} 1^{n-i}
= \frac{c}{4^n} 5^{n}
= c\left(\frac{5}{4}\right)^n.$$

#### Exercise 2

Suppose $P(X=c)=1$, then by Lemma 2.15, $P(X=c) = F(c) - F(c^-) = 1$. By Theorem 2.8, the CDF is non-decreasing and $0 \leq F(x) \leq 1$ for all $x$, therefore
$$F(x) = \begin{cases}
0, &\text{if } x < c, \\
1, &\text{if } x \geq c.
\end{cases}$$

This coincides with the CDF of a point mass distribution, hence by Theorem 2.7, this random variable has a point mass distribution, taking a single value $c$ with probability 1. Therefore, $E[X] = c$ and $V[X] = 0$.

Suppose now $V[X] = 0$ and consider two cases:
   1. $\exists c: 0 < F(c) < 1$,
   2. $\nexists c: 0 < F(c) < 1$.

Take the first case. From Theorem 2.8, $\lim_{x \to -\infty} F(x) = 0$ and therefore $\exists c_1: F(c_1) < F(c)$. Likewise, $\lim_{x \to +\infty} F(x) = 1$ implies $\exists c_2: F(c_2) > F(c)$.

By Theorem 3.10, $V[X]=0$ implies that $E[X]$ exists. Suppose $E[X] < c$, then 
$$\int_{c}^{c_2} (x - E[X])^2 dF(x) \geq \int_{c}^{c_2} (c - E[X])^2 dF(x) = (c - E[X])^2 \int_{c}^{c_2} dF(x) = (c - E[X])^2 (F(c_2) - F(c)) > 0$$
and therefore 
$$V[X] 
= \int_{-\infty}^{+\infty} (x - E[X])^2 dF(x) 
= \int_{-\infty}^{c} (x - E[X])^2 dF(x) + \int_{c}^{c_2} (x - E[X])^2 dF(x) + \int_{c_2}^{+\infty} (x - E[X])^2 dF(x)
> 0.$$

Similarly, we obtain a contradiction by assuming $E[X] > c$ and considering the integral over the interval $(c_1, c)$.

Therefore, it must be that $E[X] = c$.

From Theorem 2.8, $F$ is right-continuous, and therefore $\exists c_3 > c: F(c) \leq F(c_3) < F(c_2)$. Take 
$$\int_{c_3}^{c_2} (x - E[X])^2 dF(x) \geq \int_{c_3}^{c_2} (c_3 - c)^2 dF(x) = (c_3 - c)^2 \int_{c_3}^{c_2} dF(x) = (c_3 - c)^2 (F(c_2) - F(c_3)) > 0,$$
and thus $V[X] > 0$.

We can see that the first case leads to a contradiction. Consider now the second case: $\nexists c: 0 < F(c) < 1$, i.e. for all $x$ either $F(x)=0$ or $F(x)=1$, and from normalisation (Theorem 2.8) we know that both values are attained. Take $d = \inf \{x | F(x) = 1\}$. It exists since $\{x | F(x) = 1\}$ is bounded from below (which itself follows from the fact the any CDF is non-decreasing). From right-continuity, we see that $F(d) = 1$, and therefore $d = \min \{x | F(x) = 1\}$. This means that the CDF coincides with the CDF of a point mass distribution, and using Theorem 2.7 again, we see that it has $P(X=d)=1$.

#### Exercise 8

By linearity of expectation, 
$$E[\bar{X}_n] = E\left[ \frac{1}{n} \sum_{i=1}^n X_i \right] = \frac{1}{n} \sum_{i=1}^n E[X_i] = \frac{1}{n} \cdot n E[X_1] = \mu.$$

Since $X_1, \dots, X_n$ are i.i.d.,
$$V[\bar{X}_n] = V\left[ \frac{1}{n} \sum_{i=1}^n X_i \right] = \frac{1}{n^2} \sum_{i=1}^n V[X_i] = \frac{1}{n^2} \cdot n V[X_1] = \frac{1}{n^2} \cdot n \sigma^2 = \frac{\sigma^2}{n}.$$

By Theorem 3.15, $V[X] = E[X^2] - \mu^2$, so $E[X^2] = V[X] + \mu^2$. By Theorem 3.19, $\text{Cov}(X_i, X_j) = E[X_i X_j] - E[X_i] E[X_j]$, so $E[X_i X_j] = \text{Cov}(X_i, X_j) + E[X_i] E[X_j]$.

Since $X_1, \dots, X_n$ are i.i.d., from Theorem 3.19 also $\text{Cov}(X_i, X_j) = 0$ whenever $i \neq j$. Therefore,
$$E[X_i X_j] = \begin{cases}
\sigma^2 + \mu^2, &\text{if } i = j, \\
\mu^2, &\text{if } i \neq j.
\end{cases}$$

Consider
$$E[X_i \bar{X}_n] = E\left[X_i \frac{1}{n} \sum_{j=1}^n X_j \right] = \frac{1}{n} \sum_{j=1}^n E[X_i X_j] = \frac{\sigma^2 + n \mu^2}{n}$$
and
$$E\left[\bar{X}_n^2\right] 
= E\left[\left(\frac{1}{n}\sum_{i=1}^n X_i\right)^2\right] 
= \frac{1}{n^2} E\left[\sum_{i=1}^n\sum_{j=1}^n X_i X_j\right]
= \frac{1}{n^2} \sum_{i=1}^n\sum_{j=1}^n E[X_i X_j]
= \frac{1}{n^2} (n (\sigma^2 + \mu^2) + (n^2 - n) \mu^2)
= \frac{\sigma^2 + n \mu^2}{n}.$$

Now
$$\begin{split}
E[S_n] 
&= E\left[\frac{1}{n-1} \sum_{i=1}^n (X_i - \bar{X}_n)^2\right]
= \frac{n}{n-1} E\left[(X_i - \bar{X}_n)^2\right]
= \frac{n}{n-1} E\left[X_i^2 - 2 X_i \bar{X}_n + \bar{X}_n^2\right]
= \frac{n}{n-1} \left(E\left[X_i^2\right] - 2 E\left[X_i \bar{X}_n\right] + E\left[\bar{X}_n^2\right]\right) \\
&= \frac{n}{n-1} \left(\sigma^2 + \mu^2 - 2 \frac{\sigma^2 + n \mu^2}{n} + \frac{\sigma^2 + n \mu^2}{n}\right)
= \frac{n}{n-1} \frac{(n-1)\sigma^2}{n}
= \sigma^2.
\end{split}
$$

##### Poisson

The probability mass function is
$$f(n) = e^{-\lambda} \frac{\lambda^n}{n!}.$$

Therefore
$$E[X] 
= \sum_{n=0}^{\infty} n e^{-\lambda} \frac{\lambda^n}{n!} 
= \sum_{n=1}^{\infty} e^{-\lambda} \frac{\lambda^n}{(n-1)!}
= \lambda \sum_{n=1}^{\infty} e^{-\lambda} \frac{\lambda^{n-1}}{(n-1)!}
= \lambda \sum_{n=0}^{\infty} e^{-\lambda} \frac{\lambda^n}{n!}
= \lambda,
$$
since 
$$\sum_{n=0}^{\infty} e^{-\lambda} \frac{\lambda^n}{n!} = P(\Omega) = 1.$$

Now
$$E\left[X^2\right] 
= \sum_{n=0}^{\infty} n^2 e^{-\lambda} \frac{\lambda^n}{n!} 
= \lambda + \sum_{n=2}^{\infty} n^2 e^{-\lambda} \frac{\lambda^n}{n!} 
= \lambda + \sum_{n=2}^{\infty} e^{-\lambda} \frac{\lambda^n}{(n-2)!} 
= \lambda + \lambda^2 \sum_{n=2}^{\infty} e^{-\lambda} \frac{\lambda^{n-2}}{(n-2)!} 
= \lambda + \lambda^2 \sum_{n=0}^{\infty} e^{-\lambda} \frac{\lambda^n}{n!} 
= \lambda + \lambda^2,
$$
and so
$$V[X] = E\left[X^2\right] - E[X]^2 = \lambda + \lambda^2 - \lambda^2 = \lambda.$$

##### Exponential

The probability density function is
$$f(x) = \frac{1}{\beta} e^{-\frac{x}{\beta}},$$
where $x \geq 0$ and 0 otherwise.

Therefore
$$\begin{split}
E[X] 
&= \int_{-\infty}^{+\infty} x f(x) dx 
= \int_{0}^{+\infty} x \frac{1}{\beta} e^{-\frac{x}{\beta}} dx
= \beta \int_{0}^{+\infty} \frac{x}{\beta} e^{-\frac{x}{\beta}} d\left(\frac{x}{\beta}\right)
= \left[y = \frac{x}{\beta} \right]
= \beta \int_{0}^{+\infty} y e^{-y} dy
= \left[
  \begin{alignedat}{2}
  u       &= y             \quad & du &= dy \\
  dv &= e^{-y} dy \quad & v &= - e^{-y}
  \end{alignedat}\,
\right] \\
&= \beta \left( \left.-y e^{-y}\right|_0^{+\infty} + \int_0^{+\infty} e^{-y} dy \right)
= \beta \left.(-e^{-y})\right|_0^{+\infty}
= \beta,
\end{split}
$$

$$\begin{split}
E[X^2] 
&= \int_{-\infty}^{+\infty} x^2 f(x) dx 
= \int_{0}^{+\infty} x^2 \frac{1}{\beta} e^{-\frac{x}{\beta}} dx
= \beta^2 \int_{0}^{+\infty} \left(\frac{x}{\beta}\right)^2 e^{-\frac{x}{\beta}} d\left(\frac{x}{\beta}\right)
= \left[y = \frac{x}{\beta} \right]
= \beta^2 \int_{0}^{+\infty} y^2 e^{-y} dy
= \left[
  \begin{alignedat}{2}
  u       &= y^2             \quad & du &= 2y dy \\
  dv &= e^{-y} dy \quad & v &= - e^{-y}
  \end{alignedat}\,
\right] \\
&= \beta^2 \left( \left.-y^2 e^{-y}\right|_0^{+\infty} + 2 \int_0^{+\infty} y e^{-y} dy \right)
= 2 \beta^2 \int_0^{+\infty} y e^{-y} dy
= 2 \beta^2,
\end{split}
$$
so
$$V[X] = E[X^2] - E[X]^2 = \beta^2.$$

##### Gamma

The pdf is
$$f(x) = \frac{1}{\beta^\alpha \Gamma(\alpha)} x^{\alpha-1} e^{-\frac{x}{\beta}},$$
where $x \geq 0$ and 0 otherwise.

Then
$$E[X] = \int_{-\infty}^{+\infty} x f(x) dx 
= \int_{0}^{+\infty} x \frac{1}{\beta^\alpha \Gamma(\alpha)} x^{\alpha-1} e^{-\frac{x}{\beta}} dx
= \int_{0}^{+\infty} \frac{1}{\beta^\alpha \Gamma(\alpha)} x^{\alpha} e^{-\frac{x}{\beta}} dx
= \alpha \beta \int_{0}^{+\infty} \frac{1}{\beta^{\alpha + 1} \Gamma(\alpha + 1)} x^{\alpha} e^{-\frac{x}{\beta}} dx
= \alpha \beta,
$$
since $\frac{1}{\beta^{\alpha + 1} \Gamma(\alpha + 1)} x^{\alpha} e^{-\frac{x}{\beta}}$ is the pdf of $\text{Gamma}(\alpha + 1, \beta)$ and $\Gamma(\alpha + 1) = \alpha \Gamma(\alpha).$

Similarly,
$$E[X^2] = \int_{-\infty}^{+\infty} x^2 f(x) dx 
= \int_{0}^{+\infty} x^2 \frac{1}{\beta^\alpha \Gamma(\alpha)} x^{\alpha-1} e^{-\frac{x}{\beta}} dx
= \int_{0}^{+\infty} \frac{1}{\beta^\alpha \Gamma(\alpha)} x^{\alpha + 1} e^{-\frac{x}{\beta}} dx
= \alpha (\alpha + 1) \beta^2 \int_{0}^{+\infty} \frac{1}{\beta^{\alpha + 2} \Gamma(\alpha + 2)} x^{\alpha + 1} e^{-\frac{x}{\beta}} dx
= \alpha (\alpha + 1) \beta^2,
$$
where we used $\Gamma(\alpha + 2) = (\alpha + 1)\Gamma(\alpha + 1) = \alpha(\alpha + 1) \Gamma(\alpha)$ and the fact that
$$\frac{1}{\beta^{\alpha + 2} \Gamma(\alpha + 2)} x^{\alpha + 1} e^{-\frac{x}{\beta}}$$
is the pdf of $\text{Gamma}(\alpha + 2, \beta)$.

Finally,
$$V[X] = E[X^2] - E[X^2] = \alpha (\alpha + 1) \beta^2 - \alpha^2 \beta^2 = \alpha \beta^2.$$

##### Beta

The pdf is
$$f(x) = \frac{\Gamma(\alpha + \beta)}{\Gamma(\alpha) \Gamma(\beta)} x^{\alpha-1} (1 - x)^{\beta - 1},$$
where $0 \leq x \leq 1$ and 0 otherwise.

Then
$$
\begin{split}
E[X] 
&= \int_{-\infty}^{+\infty} x f(x) dx 
= \int_0^1 x \frac{\Gamma(\alpha + \beta)}{\Gamma(\alpha) \Gamma(\beta)} x^{\alpha-1} (1 - x)^{\beta - 1} dx
= \int_0^1 \frac{\Gamma(\alpha + \beta)}{\Gamma(\alpha) \Gamma(\beta)} x^{\alpha} (1 - x)^{\beta - 1} dx
= \left[
  \begin{alignedat}{2}
  \Gamma(\alpha + 1 + \beta) &= (\alpha + \beta) \Gamma(\alpha + \beta) \\
  \Gamma(\alpha + 1) &= \alpha \Gamma(\alpha)
  \end{alignedat}\,
\right] \\
&= \frac{\alpha}{\alpha + \beta} \int_0^1 \frac{\Gamma(\alpha + 1 + \beta)}{\Gamma(\alpha + 1) \Gamma(\beta)} x^{\alpha} (1 - x)^{\beta - 1} dx
= \frac{\alpha}{\alpha + \beta},
\end{split}
$$
since
$$\frac{\Gamma(\alpha + 1 + \beta)}{\Gamma(\alpha + 1) \Gamma(\beta)} x^{\alpha} (1 - x)^{\beta - 1}$$
is the pdf of $\text{Beta}(\alpha + 1, \beta)$.

Similarly,
$$
E[X^2] 
= \int_{-\infty}^{+\infty} x^2 f(x) dx 
= \int_0^1 x^2 \frac{\Gamma(\alpha + \beta)}{\Gamma(\alpha) \Gamma(\beta)} x^{\alpha-1} (1 - x)^{\beta - 1} dx
= \int_0^1 \frac{\Gamma(\alpha + \beta)}{\Gamma(\alpha) \Gamma(\beta)} x^{\alpha + 1} (1 - x)^{\beta - 1} dx
= \frac{\alpha(\alpha + 1)}{(\alpha + \beta)(\alpha + \beta + 1)},
$$
therefore
$$V[X] = E[X^2] - E[X]^2 = \frac{\alpha(\alpha + 1)}{(\alpha + \beta)(\alpha + \beta + 1)} - \frac{\alpha^2}{(\alpha + \beta)^2}
= \frac{\alpha\beta}{(\alpha + \beta)^2 (\alpha + \beta + 1)}.$$

##### Antithetic variates

Suppose $U \sim \text{Uniform}(0, 1)$, then also $1 - U \sim \text{Uniform}(0, 1)$.

From the results of Exercise 15 in Chapter 2,
$$X_1 = -\beta \log(1 - U) \sim \text{Exp}(\beta)$$
and
$$X_2 = -\beta \log U \sim \text{Exp}(\beta).$$

We find the CDF of $Y = X_1 + X_2$.

$$
P(X_1 + X_2 \leq y) 
= P(-\beta \log(1 - U) - \beta \log U \leq y)
= P(\log(U(1 - U)) \geq -\beta y)
= P(U(1 - U) \geq e^{-\beta y}).
$$

Solving the equation $u^2 - u + e^{-\beta y} \leq 0$, we have 
$$u_{1,2} = \frac{1 \pm \sqrt{1 - 4 e^{-\beta y}}}{2},$$
and therefore
$$P(U(1 - U) \geq e^{-\beta y}) = \sqrt{1 - 4e^{-\beta y}},$$
as long as $e^{-\beta y} \leq y$, which only holds when $y \geq \frac{1}{\beta} \log 4$.

Therefore the cdf of $Y$ is
$$
F(y) = \begin{cases}
0, &\text{if } y \leq \frac{\log 4}{\beta}, \\
\sqrt{1 - 4e^{-\beta y}}, &\text{if } y > \frac{\log 4}{\beta}.
\end{cases}
$$

We can thus also find the pdf of $Y$:
$$
f(y) = \frac{dF(y)}{dy} = \begin{cases}
0, &\text{if } y \leq \frac{\log 4}{\beta}, \\
\frac{2\beta e^{-\beta y}}{\sqrt{1 - 4e^{-\beta y}}}, &\text{if } y > \frac{\log 4}{\beta}.
\end{cases}
$$

We can see that in this case, $Y$ does not have a Gamma distribution either.

##### Independent variables

If $X_1, \dots, X_n$ are independent, and $Y = \sum_{i=1}^n X_i$, then by Lemma 3.31,
$$\psi_Y(t) = \prod_{i=1}^n \psi_i(t) = \frac{1}{(1-\beta t)^n}.$$

This is identical to the expression for the mdf of the $\text{Gamma}(n, \beta)$, therefore by Theorem 3.33, $Y \sim \text{Gamma}(n, \beta)$.
