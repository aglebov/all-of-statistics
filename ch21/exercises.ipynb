## Exercise 1

The density estimator is given by expression (21.14):
$$\hat{f}(x) = \sum_{j=1}^J \hat{\beta}_j \phi_j(x),$$
thus the risk is given by
$$\begin{split}
R(f,\hat{f}) 
&= \mathbb{E}\left[ \int_0^1 \left(f(x) - \hat{f}(x)\right)^2 dx \right] \\
&= \mathbb{E}\left[ \int_0^1 \left(\sum_{j=1}^\infty \beta_j \phi_j(x) - \sum_{j=1}^J \hat{\beta}_j \phi_j(x) \right)^2 dx \right] \\
&= \mathbb{E}\left[ \int_0^1 \left(\sum_{j=1}^J \beta_j \phi_j(x) + \sum_{j=J+1}^\infty \beta_j \phi_j(x) - \sum_{j=1}^J \hat{\beta}_j \phi_j(x) \right)^2 dx \right] \\
&= \mathbb{E}\left[ \int_0^1 \left(\sum_{j=1}^J (\beta_j - \hat{\beta}_j) \phi_j(x) + \sum_{j=J+1}^\infty \beta_j \phi_j(x) \right)^2 dx \right] \\
&= \mathbb{E}\left[ \int_0^1 \left( \left(\sum_{j=1}^J (\beta_j - \hat{\beta}_j) \phi_j(x) \right)^2 + 2 \sum_{j=1}^J (\beta_j - \hat{\beta}_j) \phi_j(x) \sum_{k=J+1}^\infty \beta_k \phi_k(x) + \left( \sum_{j=J+1}^\infty \beta_j \phi_j(x) \right)^2 \right) dx \right] \\
&= \mathbb{E}\left[ \int_0^1 \left( \sum_{j=1}^J \sum_{k=1}^J (\beta_j - \hat{\beta}_j) (\beta_k - \hat{\beta}_k) \phi_j(x) \phi_k(x) + 2 \sum_{j=1}^J (\beta_j - \hat{\beta}_j) \phi_j(x) \sum_{k=J+1}^\infty \beta_k \phi_k(x) + \sum_{j=J+1}^\infty \sum_{k=J+1}^\infty \beta_j \beta_k \phi_j(x) \phi_k(x) \right) dx \right] \\
&= \mathbb{E}\left[ \sum_{j=1}^J \sum_{k=1}^J (\beta_j - \hat{\beta}_j) (\beta_k - \hat{\beta}_k) \int_0^1 \phi_j(x) \phi_k(x) dx + 2 \sum_{j=1}^J \sum_{k=J+1}^\infty (\beta_j - \hat{\beta}_j) \beta_k \int_0^1 \phi_j(x) \phi_k(x) dx + \sum_{j=J+1}^\infty \sum_{k=J+1}^\infty \beta_j \beta_k  \int_0^1 \phi_j(x) \phi_k(x) dx \right] \\
&= \mathbb{E}\left[ \sum_{j=1}^J (\beta_j - \hat{\beta}_j)^2 + \sum_{j=J+1}^\infty  \beta_j^2 \right] \\
&=  \sum_{j=1}^J \mathbb{E}\left[ (\beta_j - \hat{\beta}_j)^2 \right] + \sum_{j=J+1}^\infty  \beta_j^2  \\
&=  \sum_{j=1}^J \mathbb{V}[\hat{\beta}_j] + \sum_{j=J+1}^\infty  \beta_j^2  \\
&=  \sum_{j=1}^J \frac{\sigma_j}{n} + \sum_{j=J+1}^\infty  \beta_j^2  \\
\end{split}$$

We show that $\hat{\sigma}_j^2$ is an unbiased estimator of $\sigma_j^2$:
$$\begin{split}
\mathbb{E}[\hat{\sigma}_j^2]
&= \mathbb{E}\left[ \frac{1}{n-1} \sum_{i=1}^n \left(\phi_j(X_i) - \hat{\beta}_j\right)^2  \right] \\
&= \mathbb{E}\left[ \frac{1}{n-1} \sum_{i=1}^n \left(\phi_j(X_i) - \frac{1}{n} \sum_{k=1}^n \phi_j(X_k) \right)^2  \right] \\
&= \frac{1}{n-1} \mathbb{E}\left[ \sum_{i=1}^n \left(\phi_j(X_i)^2 - 2 \phi_j(X_i) \frac{1}{n} \sum_{k=1}^n \phi_j(X_k) + \frac{1}{n^2} \sum_{k=1}^n \sum_{l=1}^n \phi_j(X_k) \phi_j(X_l) \right) \right] \\
&= \frac{1}{n-1} \mathbb{E}\left[ \sum_{i=1}^n \phi_j(X_i)^2 - \frac{2}{n}  \sum_{i=1}^n \phi_j(X_i) \sum_{k=1}^n \phi_j(X_k) +  \sum_{i=1}^n \frac{1}{n^2} \sum_{k=1}^n \sum_{l=1}^n \phi_j(X_k) \phi_j(X_l) \right] \\
&= \frac{1}{n-1} \mathbb{E}\left[ \sum_{i=1}^n \phi_j(X_i)^2 - \frac{2}{n}  \sum_{i=1}^n \sum_{k=1}^n \phi_j(X_i) \phi_j(X_k) + \frac{1}{n} \sum_{k=1}^n \sum_{l=1}^n \phi_j(X_k) \phi_j(X_l) \right] \\
&= \frac{1}{n-1} \mathbb{E}\left[ \sum_{i=1}^n \phi_j(X_i)^2 - \frac{1}{n}  \sum_{i=1}^n \sum_{k=1}^n \phi_j(X_i) \phi_j(X_k)  \right] \\
&= \frac{1}{n-1} \left( \sum_{i=1}^n  \mathbb{E}\left[ \phi_j(X_i)^2 \right] - \frac{1}{n} \sum_{i=1}^n \sum_{k=1}^n  \mathbb{E}\left[ \phi_j(X_i) \phi_j(X_k) \right]  \right) \\
&= \frac{1}{n-1} \left( n \mathbb{E}\left[ \phi_j(X_i)^2 \right] - \frac{1}{n}\left( \sum_{i=1}^n \mathbb{E}\left[ \phi_j(X_i)^2 \right] + \sum_{i=1}^n \sum_{k\neq i}^n \mathbb{E}\left[ \phi_j(X_i) \right] \mathbb{E}\left[ \phi_j(X_j) \right]  \right) \right) \\
&= \frac{1}{n-1} \left( n \mathbb{E}\left[ \phi_j(X_i)^2 \right] - \frac{1}{n}\left( n \mathbb{E}\left[ \phi_j(X_i)^2 \right] + n(n-1) \mathbb{E}\left[ \phi_j(X_i) \right]^2 \right) \right) \\
&= \frac{1}{n-1} \left( n \mathbb{E}\left[ \phi_j(X_i)^2 \right] - \mathbb{E}\left[ \phi_j(X_i)^2 \right] - (n-1) \mathbb{E}\left[ \phi_j(X_i) \right]^2  \right) \\
&= \frac{1}{n-1} \left( (n - 1) \mathbb{E}\left[ \phi_j(X_i)^2 \right] - (n-1) \mathbb{E}\left[ \phi_j(X_i) \right]^2  \right) \\
&= \mathbb{E}\left[ \phi_j(X_i)^2 \right] - \mathbb{E}\left[ \phi_j(X_i) \right]^2  \\
&= \mathbb{E}\left[ \phi_j(X_i)^2 \right] - \mathbb{E}\left[ \phi_j(X_i) \right]^2  \\
&= \mathbb{E}\left[ \left( \phi_j(X_i) - \mathbb{E}[\phi_j(X_i)]\right)^2 \right] \\
&= \mathbb{E}\left[ \left( \phi_j(X_i) - \beta_j\right)^2 \right] \\
&= \sigma_j^2,
\end{split}$$
since
$$\mathbb{E}[\phi_j(X_i)] = \int_0^1 \phi_j(x) f(x) dx = \beta_j.$$

Finally,
$$\frac{\sigma_j^2}{n} = \mathbb{V}[\hat{\beta}_j] = \mathbb{E}[\hat{\beta}_j^2] - \mathbb{E}[\hat{\beta}_j]^2 = \mathbb{E}[\hat{\beta}_j] - \beta_j^2,$$
so
$$\beta_j^2 = \mathbb{E}[\hat{\beta}_j^2] - \frac{\sigma_j^2}{n} = \mathbb{E}\left[\hat{\beta}_j^2 - \frac{\hat{\sigma}_j^2}{n}\right],$$
thus $\hat{\beta}_j^2 - \frac{\hat{\sigma}_j^2}{n}$ is an unbiased estimator of $\beta_j^2$.

```python
def cosine(j, x):
    return np.where(j > 0, np.sqrt(2) * np.cos(j * np.pi * x), 1)
```

```python
def fn_hat_cosine(gen_betas):
    def func1(n):
        j = np.arange(n + 1).reshape(1, -1)
        betas = gen_betas(j).reshape(1, -1)
        def func2(x):
            x = x.reshape(-1, 1)
            return np.sum(betas * cosine(j, x), axis=1)
        return func2
    return func1
```

```python
x = data['Al'].values
y = data['RI'].values
sort_idx = np.argsort(x)
x = x[sort_idx]
y = y[sort_idx]
n = x.shape[0]
```

```python
def rescale(xs, x):
    return (xs - np.min(x)) / (np.max(x) - np.min(x))
```

```python
def approx_beta_reg(x, y):
    x = rescale(x, x).reshape(-1, 1)
    j = np.arange(n + 1).reshape(1, -1)
    return np.mean(y.reshape(-1, 1) * cosine(j, x), axis=0)
```

```python
def estimate_sigma_sq_hat(beta_hat):
    n = beta_hat.shape[0]
    k = n // 4
    return n / k * np.sum(beta_hat[-k:] ** 2)
```

```python
def estimate_risk(beta_hat, J):
    sigma_sq_hat = estimate_sigma_sq_hat(beta_hat)
    beta_sq_hat = np.maximum(beta_hat ** 2 - sigma_sq_hat / n, 0)
    return (np.sum(beta_sq_hat) - np.cumsum(beta_sq_hat))[:-1] + J * sigma_sq_hat / n
```

```python
beta_hat = approx_beta_reg(x, y)
J = np.arange(1, n + 1)
R = estimate_risk(beta_hat, J)
```

```python
fig, ax = plt.subplots()
ax.plot(J, R);
ax.set_xlabel('$J$');
ax.set_ylabel('$R(J)$');
ax.set_title('Risk estimate');
```

The risk is minimised when $J$ is

```python
J_opt = J[np.argmin(R)]
J_opt
```

```python
def fit_cosine(x, y, J, alpha=0.05):
    beta_hat = approx_beta_reg(x, y)
    sigma_sq_hat = estimate_sigma_sq_hat(beta_hat)
    n = x.shape[0]
    j = np.arange(J).reshape(1, -1)
    def func(xs):
        xs = rescale(xs, x).reshape(-1, 1)
        phi_vals = cosine(j, xs)
        r_hat = np.sum(beta_hat[:J].reshape(1, -1) * phi_vals, axis=1)
        a = np.sqrt(np.sum(phi_vals ** 2, axis=1))
        c = a * np.sqrt(sigma_sq_hat * stats.chi2.ppf(1 - alpha, df=J) / n)
        return r_hat, r_hat - c, r_hat + c
    return func
```

## Exercise 8

Note that 
$$\phi^2(x) = \psi^2(x) = \begin{cases}
1, &\text{ if } 0 \leq x < 1 \\
0, &\text{ otherwise }
\end{cases}$$
thus
$$\int_0^1 \phi^2(x) dx = \int_0^1 dx = 1$$
and for any $j \geq 0$, $0 \leq k \leq 2^j - 1$,
$$\begin{split}
\int_0^1 \psi_{j,k}^2(x) dx 
&= \int_0^1 \left( 2^{j/2} \psi(2^j x - k) \right)^2 dx \\
&= 2^j \int_0^1 \psi^2(2^j x - k) dx \\
&= \int_{-k}^{2^j-k} \psi^2(2^j x - k) d(2^j x - k) \\
&= \int_{-k}^{2^j-k} \psi^2(y) dy \\
&= \int_0^1 \psi^2(y) dy \\
&= 1.
\end{split}$$

For any $j \geq 0$, $0 \leq k \leq 2^j - 1$, we have
$$\begin{split}
\int_0^1 \psi_{j,k}(x) dx 
&= \int_0^1 2^{j/2} \psi(2^j x - k) dx \\
&= 2^{-j/2} \int_{-k}^{2^j - k} \psi(2^j x - k) d(2^j x - k) \\
&= 2^{-j/2} \int_{-k}^{2^j - k} \psi(y) dy \\
&= 2^{-j/2} \int_0^1 \psi(y) dy \\
&= 0, \\
\end{split}$$
therefore
$$\int_0^1 \phi(x) \psi_{j,k}(x) dx = \int_0^1 \psi_{j,k}(x) dx = 0.$$

For any $j \geq 0$, $0 \leq k, l \leq 2^j - 1$ and $k \neq l$, we have
$$\begin{split}
\psi_{j,k}(x) \neq 0 &\iff 0 \leq 2^j x - k \leq 1 \iff \frac{k}{2^j} \leq x \leq \frac{k+1}{2^j} \\
\psi_{j,l}(x) \neq 0 &\iff 0 \leq 2^j x - l \leq 1 \iff \frac{l}{2^j} \leq x \leq \frac{l+1}{2^j} \\
\end{split}$$
The intervals where $\psi_{j,k}(x) \neq 0$ and $\psi_{j,l}(x) \neq 0$ intersect in at most a single point, so $\psi_{j,k}(x) \psi_{j,l}(x)$ is non-zero for at most one point, and thus
$$\int_0^1 \psi_{j,k}(x) \psi_{j,l}(x) dx = 0.$$

Let $0 \leq i < j$ and $0 \leq k \leq 2^i - 1$, $0 \leq l \leq 2^j - 1$. Then
$$\psi_{j,l}(x) \neq 0 \iff 0 \leq 2^j x - l \leq 1 \iff \frac{l}{2^j} \leq x \leq \frac{l+1}{2^j}$$
and
$$\begin{split}
\psi_{i,k}(x) &= \begin{cases}
-1, &\text{if } \frac{k}{2^i} \leq x \leq \frac{k + \frac{1}{2}}{2^i} \\
1, &\text{if } \frac{k + \frac{1}{2}}{2^i} < x \leq \frac{k + 1}{2^i} \\
0, &\text{otherwise}
\end{cases} \\
\end{split}$$

We show that $\psi_{i,k}(x)$ is constant on the interval where $\psi_{j,l}(x) \neq 0$ except possibly at a single point.

Suppose $l < 2^{j-i} k$, then $l + 1 \leq 2^{j-i} k$, so
$$\frac{l}{2^j} < \frac{l+1}{2^j} \leq \frac{k}{2^i}$$
and $\psi_{i,k}(x) = 0$ for all $\frac{l}{2^j} \leq x \leq \frac{l+1}{2^j}$ except possibly at $x = \frac{l+1}{2^j}$.

Suppose $2^{j-i} k \leq l < 2^{j-i} \left(k + \frac{1}{2}\right)$, then $l + 1 \leq 2^{j-i} \left(k + \frac{1}{2}\right)$ and
$$\frac{k}{2^i} \leq \frac{l}{2^j} < \frac{l+1}{2^j} \leq \frac{k + \frac{1}{2}}{2^i},$$
so $\psi_{i,k}(x) = -1$ for all $\frac{l}{2^j} \leq x \leq \frac{l+1}{2^j}$.

Suppose $2^{j-i} \left(k + \frac{1}{2}\right) \leq l < 2^{j-i} (k + 1) $, then $l + 1 \leq 2^{j-i} (k + 1)$ and
$$\frac{k + \frac{1}{2}}{2^i} \leq \frac{l}{2^j} < \frac{l+1}{2^j} \leq \frac{k + 1}{2^i},$$
so $\psi_{i,k}(x) = 1$ for all $\frac{l}{2^j} \leq x \leq \frac{l+1}{2^j}$ except possibly at $\frac{l}{2^j}$.

Suppose $2^{j-i}(k + 1) \leq l $, then $l + 1 > 2^{j-i} (k + 1)$ and
$$\frac{k +1}{2^i} \leq \frac{l}{2^j} < \frac{l+1}{2^j},$$
so $\psi_{i,k}(x) = 0$ for all $\frac{l}{2^j} \leq x \leq \frac{l+1}{2^j}$ except possibly at $\frac{l}{2^j}$.

Then
$$\begin{split}
\int_0^1 \psi_{i,k}(x) \psi_{j,l}(x) dx 
&= \int_{\frac{l}{2^j}}^{\frac{l + 1}{2^j}} \psi_{i,k}(x) \psi_{j,l}(x) dx \\
&= \psi_{i,k}\left(\frac{l + \frac{1}{2}}{2^j}\right) \int_{\frac{l}{2^j}}^{\frac{l + 1}{2^j}} \psi_{j,l}(x) dx \\
&= 0.
\end{split}$$

We conclude that $\{\phi, W_0, W_1, \dots\}$ are orthonormal.

We use Definition 20.5 to obtain cross-validation estimator of risk. We evaluate $\int \left(\hat{f}(x)\right)^2 dx$ numerically by partitioning the interval into sub-intervals of size $2^{-J}$ and evaluating the $\left(\hat{f}(x)\right)^2$ in the middle of each subinterval.

```python
def cross_validation_risk(J):
    f_hat = fit_wavelet(J, x, np.ones(n))
    bounds = np.linspace(0, 1, 2 ** J + 1)
    xs = (bounds[1:] + bounds[:-1]) / 2
    t1 = np.sum(f_hat(xs) ** 2) / 2 ** J
    t2 = 2 * np.mean(
        np.fromiter((fit_wavelet(J, np.delete(x, i), np.ones(n - 1))(x[i])[0] for i in range(n)), float)
    )
    return t1 - t2
```
