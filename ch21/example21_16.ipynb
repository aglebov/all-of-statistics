```python
from math import floor

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
from wavelet import haar_mother, haar_father, phi, psi, fit_wavelet
```

```python
rng = np.random.default_rng(12345)
```

```python
def doppler(x):
    return np.sqrt(x * (1 - x)) * np.sin(2.1 * np.pi / (x + 0.05))
```

```python
n = 2048
```

```python
x = np.linspace(0, 1, n)
eps = stats.norm.rvs(scale=0.1, size=n, random_state=rng)
y = doppler(x) + eps
```

```python
J = floor(np.log2(n))
```

```python
xs = np.linspace(np.min(x), np.max(x), 500)
r_hat = fit_wavelet(J, x, y)(xs)
```

```python
fig, ax = plt.subplots()
ax.scatter(x, y, alpha=0.2);
ax.plot(xs, r_hat, color='red');
```
